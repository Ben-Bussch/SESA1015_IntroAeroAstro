$$
L = \frac{1}{2} \rho V^2 S C_L
D = \frac{1}{2} \rho V^2 S C_D
$$

```python
def compute_lift_force(V, S, CL, rho):
    
    #
    L = 1/2*(rho*V**2*S*CL)
    
    #
    return L

def compute_drag_force(V, S, CD, rho):
    
    #
    D = 1/2*(rho*V**2*S*CD)
    
    #
    return D
```

```python
g = 9.81
rho = 1.225
W = 2.5*g
V = 18
b = 1.8
AR = 8.2
S = (b**2)/AR
CL = W/(0.5*rho*(V**2)*S)
CD = 0.099
Lift = compute_lift_force(V, S, CL, rho)
Drag = compute_drag_force(V, S, CD, rho)
print("The aircraft's lift and drag forces are L = {:.2f} N and D = {:.2f} N, respectively.".format(Lift, Drag))
```

Study the change in lift force and drag force with flight speed

```python
velocity_vector = np.array([8,12,16,20,24,28,32])
```

```python
lift_force_vector = compute_lift_force(velocity_vector, S, CL, rho)
drag_force_vector = compute_drag_force(velocity_vector, S, CD, rho)
print("The computed lift force vector is {} N".format(lift_force_vector))
print("The computed drag force vector is {} N".format(drag_force_vector))
```

Plot the data you obtained

```python
import matplotlib.pyplot as plt
```

```python
plt.figure
plt.plot(velocity_vector,lift_force_vector,'.-',label="Lift force")
plt.plot(velocity_vector,drag_force_vector,'.-',label="Drag force")
plt.xlabel("Flight speed [m/s]")
plt.ylabel("Aerodynamic force [N]")
plt.legend()
plt.grid()
plt.show()
```
